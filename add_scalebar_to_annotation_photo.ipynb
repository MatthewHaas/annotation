```python
# Load necessary packages
import imageio
import os
import fiona
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar
from matplotlib_scalebar.scalebar import SI_LENGTH
```

```python
# Set working directory
os.chdir("/Users/matthewwilliamhaas/Documents/wild_rice/Figures")
```

```python
plt.figure()
im = plt.imread('tissues_for_annotation_original.jpg') # Read in image
plt.imshow(im)
```

```python
# Define scale bar
scalebar = ScaleBar(dx=80, 
                    units="um", # pixel size 
                    dimension=SI_LENGTH, 
                    location="lower left", # positions the scale bar (default=upper right)
                    box_alpha=0, # Makes box transparent
                    color="white", # Makes font and scale bar white
                    font_properties="Times-10") # Makes label Times New Roman (serif) and size=30
```

```python
# Add scale bar
plt.gca().add_artist(scalebar)
plt.axis("off") # turn off the axis
fig = plt.imshow(im)

plt.savefig("tissues_for_annotation_with_scalebar.jpg", dpi=600) # save figure to a file
```
